#### Ex.3. For one dimensional distributions plot histograms. Assign bin values to represent the geometry of distribution. Compare the histograms (normalized) with plots of analytical formulas for distributions (see eg. Wikipedia).

##### a) Uniform distribution

```python
n_bins = 100
n, bins, patches = plt.hist(uniform_distr, n_bins, density=True)
plt.plot(bins, np.ones(n_bins+1))
plt.title("Uniform distribution")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Uniform distribution", "Analytical formula"])
plt.show()
```

Analytical formula for uniform distribution is $\frac{1}{b-a}$, so on the interval $[0,1]$ it's a constant function $f(x) = 1$. 

##### b) Normal distribution with mean $\mu = 0$ and standard deviation $\sigma = 1$.

```python
normal_distr = lambda u, sigma, x: (1/(sigma*np.sqrt(2*np.pi)))*np.exp((-(x-u)**2)/(2*(sigma**2)))

n_bins = 100
n, bins, patches = plt.hist(normal_distr_dev_1, n_bins, density=True)
plt.plot(bins, normal_distr(0,1,bins))
plt.title("Normal distribution  $\mu = 0$, $\sigma = 1$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Normal distribution", "Analytical formula"])
plt.show()
```

```python
n_bins = 100
n, bins, patches = plt.hist(normal_distr_dev_10, n_bins, density=True)
plt.plot(bins, normal_distr(0,10,bins))
plt.title("Normal distribution  $\mu = 0$, $\sigma = 10$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["Normal distribution", "Analytical formula"])
plt.show()
```
